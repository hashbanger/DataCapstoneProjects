```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
sns.set(style = 'darkgrid', palette = 'pastel')
```

```python
df = pd.read_excel('DrinkingWater.xlsx')
```

```python
fig, axes = plt.subplots(5,2)
```

```python
plt.figure(figsize= (10,20))
for i in range(0,len(df.columns[2:-1])):
    ax = plt.subplot(5,2,i+1)
    sns.barplot(df['State/UT'], df[df.columns[2:-1][i]], ax = ax)
    plt.xticks(rotation = 90)
    plt.show()
```

```python
plt.figure(figsize= (20,20))
plt.subplot(5,2,1)
sns.barplot(df['State/UT'], df['Primary_Only'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,2)
sns.barplot(df['State/UT'], df['Primary_with_U_Primary'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,3)
sns.barplot(df['State/UT'], df['Primary_with_U_Primary_Sec'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,4)
sns.barplot(df['State/UT'], df['Primary_with_U_Primary_Sec_HrSec'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,5)
sns.barplot(df['State/UT'], df['U_Primary_Only'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,6)
sns.barplot(df['State/UT'], df['U_Primary_With_Sec'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,7)
sns.barplot(df['State/UT'], df['U_Primary_With_Sec_HrSec'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,8)
sns.barplot(df['State/UT'], df['Sec_Only'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,9)
sns.barplot(df['State/UT'], df['Sec_with_HrSec.'], palette = 'winter')
plt.xticks(rotation = 90)
plt.subplot(5,2,10)
sns.barplot(df['State/UT'], df['HrSec_Only'], palette = 'winter')
plt.xticks(rotation = 90)
plt.show()
```

From the above plots we can say that there's not a major Drinking Water problem but states like *Meghalaya and Nagaland* have comparatively less percentage of availablility. We neglect *Mizoram* as it has mostly **0s** which are likely to be missing values.

```python
print(df[df['State/UT'] == 'Meghalaya'].mean(axis = 1), end = '\n\n')
print(df[df['State/UT'] == 'Nagaland'].mean(axis = 1))
```
