```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import shapefile as shp
```

```python
sns.set_style('darkgrid')
sns.set_palette('pastel')
```

### Importing the Data

```python
df = pd.read_excel('GrossEnrollment.xlsx')
```

```python
df.head(10)
```

First we take a look at the missing values in the dataset.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(), cbar= False)
plt.show()
```

So we have quite a few missing values, these are some in **Male** column and some in **Female** columns for varying states so rather than dropping we can consider filling it with 0s

```python
df.fillna(value= 0, inplace= True)
```

### Total Enrollment Distribution

```python
plt.figure(figsize=(20,10))
sns.kdeplot(df['ALL CATEGORIES - MALE'],color= 'blue', shade = True)
sns.kdeplot(df['ALL CATEGORIES - FEMALE'],color= 'red', shade = True)
plt.show()
```

#### Comparision between Male and Female Enrollment in each state

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(y = df['STATES/UTs'], x = df['ALL CATEGORIES - MALE'], color= 'blue', alpha = 0.7)

plt.subplot(1,2,2)
sns.barplot(y = df['STATES/UTs'], x = df['ALL CATEGORIES - FEMALE'], color= 'red', alpha = 0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13,10))
sns.barplot(y = df['STATES/UTs'], x = df['ALL CATEGORIES - MALE'], color= 'blue' ,alpha = 0.5)
sns.barplot(y = df['STATES/UTs'], x = df['ALL CATEGORIES - FEMALE'], color= 'red' ,alpha = 0.5)
plt.show()
```

The bars having **blue** peaks shows higher male students enrolled while the **pink** peaks denote more females.

We can compare the enrollment **SC** students and **ST** students for each state.

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(y = df['STATES/UTs'], x = df['SC - MALE'], color= 'blue' ,alpha = 0.5)
sns.barplot(y = df['STATES/UTs'], x = df['SC - FEMALE'], color= 'red' ,alpha = 0.5)

plt.subplot(1,2,2)
sns.barplot(y = df['STATES/UTs'], x = df['ST - MALE'], color= 'blue' ,alpha = 0.5)
sns.barplot(y = df['STATES/UTs'], x = df['ST - FEMALE'], color= 'red' ,alpha = 0.5)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(y = df['STATES/UTs'], x = df['SC - TOTAL'], color= 'blue' ,alpha = 0.5)
plt.subplot(1,2,2)
sns.barplot(y = df['STATES/UTs'], x = df['ST - TOTAL'], color= 'black' ,alpha = 0.5)
plt.tight_layout()
plt.show()
```
