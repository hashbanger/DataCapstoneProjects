```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Electricity.xlsx')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(), cbar = False)
plt.show()
```
