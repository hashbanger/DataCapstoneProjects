```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
sns.set(style = 'darkgrid' ,palette = 'pastel')
```

```python
df = pd.read_excel('DropOutRate.xlsx')
```

```python
df.head(10)
```

```python
sns.heatmap(df.isnull(), cbar = False)
plt.show()
```

```python
pd.DataFrame.replace(df, to_replace= 'NR', value= 0, inplace= True)
pd.DataFrame.replace(df, to_replace= 'Uppe_r_Primary', value= 0, inplace= True)
```

```python
df.columns[2:]
```

```python
for col in df.columns[2:]:
    df[col] = df[col].astype('float')
```

```python
df.info()
```

```python
plt.figure(figsize= (18, 17))

plt.subplot(2,1,1)
sns.barplot(df['State_UT'], df['Primary_Boys'], palette= 'Blues')
plt.xticks(rotation = 90)

plt.subplot(2,1,2)
sns.barplot(df['State_UT'], df['Primary_Girls'], palette= 'Reds')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

It can be observed that there has been subsequently higher dropout rates in **Primary** in the *Eastern India* which also dropped in **2014-15**

```python
plt.figure(figsize= (18, 17))

plt.subplot(2,1,1)
sns.barplot(df['State_UT'], df['Upper Primary_Boys'], hue= df['Year'], palette= 'Blues')
plt.xticks(rotation = 90)

plt.subplot(2,1,2)
sns.barplot(df['State_UT'], df['Upper Primary_Girls'], hue = df['Year'], palette= 'Reds')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (20, 15))

plt.subplot(2,2,1)
sns.barplot(df['State_UT'], df['Primary_Total'], hue= df['Year'], palette= 'Greens')
plt.xticks(rotation = 90)
plt.yticks([int(i) for i in range(5,55,5)])

plt.subplot(2,2,2)
sns.barplot(df['State_UT'], df['Upper Primary_Total'], hue = df['Year'], palette= 'Reds')
plt.xticks(rotation = 90)
plt.yticks([int(i) for i in range(5,55,5)])

plt.subplot(2,2,3)
sns.barplot(df['State_UT'], df['Secondary _Total'], hue = df['Year'], palette= 'Blues')
plt.xticks(rotation = 90)
plt.yticks([int(i) for i in range(5,55,5)])

plt.subplot(2,2,4)
sns.barplot(df['State_UT'], df['HrSecondary_Total'], hue = df['Year'], palette= 'Greys')
plt.xticks(rotation = 90)
plt.yticks([int(i) for i in range(5,55,5)])

plt.tight_layout()
plt.show()
```

As it can be observed that the **drop out** rates in the **Secondary** and **Hr. Secondary** higher than of **Primary** and **Upper Primary**. 

```python
temp = df.T[1:].T
for col in temp.columns[1:]:
    temp[col] = temp[col].astype('float')
temp = temp.groupby(['Year']).mean()
temp.drop(labels= ['Primary_Boys', 'Primary_Girls','Upper Primary_Boys', 'Upper Primary_Girls'
                  ,'Secondary _Boys', 'Secondary _Girls','HrSecondary_Boys', 'HrSecondary_Girls'], axis = 1, inplace = True)
temp.plot(kind = 'bar', stacked = True, figsize = (11,7))
plt.legend(loc = 3,bbox_to_anchor=(1, 0.5))
plt.show()
```

We can see that each year the **drop out** rates are always highest for **Secondary** students.

```python
plt.figure(figsize = (15,8))
df['BoysTotal'] = (df['Primary_Boys'] + df['Upper Primary_Boys'] + df['Secondary _Boys'] + df['HrSecondary_Boys'])/4
df['GirlsTotal'] = (df['Primary_Girls'] + df['Upper Primary_Girls'] + df['Secondary _Girls'] + df['HrSecondary_Girls'])/4
sns.barplot(df['State_UT'], df['BoysTotal']-df['GirlsTotal'], palette = 'inferno')
plt.xticks(rotation = 90)
plt.yticks([int(i) for i in range(-4,5,1)])
plt.ylabel('More Girls    |    More Boys')
plt.show()
```

As visible there are more **Boys** droputs overall than **Girls**
