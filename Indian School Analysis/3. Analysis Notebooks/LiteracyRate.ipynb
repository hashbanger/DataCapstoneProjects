```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
sns.set(style = 'darkgrid', palette = 'pastel')
```

```python
df = pd.read_excel('LiteracyRate.xlsx')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(), cbar = False)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.kdeplot(df['Literacy Rate (Persons) - Total - 2001'], color = 'red', shade = True)
sns.kdeplot(df['Literacy Rate (Persons) - Total - 2011'], color = 'green', shade = True)
plt.legend(['2001','2011'],loc = 1)
plt.axvline(x = df['Literacy Rate (Persons) - Total - 2001'].mean(), color = 'red')
plt.axvline(x = df['Literacy Rate (Persons) - Total - 2011'].mean(), color = 'green')
plt.show()
```

As it is clearly visible that the **Literacy Rate** has increased overall in the decade of 2001-2011.  
The overall mean has increased from 70% to 78%

```python
plt.figure(figsize=(10,10))
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= df['Literacy Rate (Persons) - Rural - 2001'], alpha = 0.5, color = 'red')
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= df['Literacy Rate (Persons) - Rural - 2011'], alpha = 0.5, color = 'green')
plt.ylabel('States')
plt.xlabel('Literacy Rate')
plt.title('Literacy Rate Comparision in 2001 and 2011 in Rurals', fontdict = {'fontsize':15})
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= df['Literacy Rate (Persons) - Urban - 2001'], alpha = 0.5, color = 'red')
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= df['Literacy Rate (Persons) - Urban - 2011'], alpha = 0.5, color = 'blue')
plt.ylabel('States')
plt.xlabel('Literacy Rate')
plt.title('Literacy Rate Comparision in 2001 and 2011 in Urbans', fontdict = {'fontsize':15})
plt.show()
```

```python
ruralIncreases = df['Literacy Rate (Persons) - Rural - 2011'] - df['Literacy Rate (Persons) - Rural - 2001']
urbanIncreases = df['Literacy Rate (Persons) - Urban - 2011'] - df['Literacy Rate (Persons) - Urban - 2001']
```

```python
plt.figure(figsize=(20,10))
plt.suptitle('Literacy Increase Comparision between Rural and Urban')

plt.subplot(1,2,1)
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= ruralIncreases, alpha = 0.6, palette = 'copper')
plt.xticks([int(i) for i in range(0,18,2)])
plt.ylabel('States')
plt.xlabel('Literacy Rate')
plt.title('Literacy Rate Increase in Rurals')

plt.subplot(1,2,2)
sns.barplot(y = df['Country/ States/ Union Territories Name'], x= urbanIncreases, alpha = 0.6, palette = 'winter')
plt.xticks([int(i) for i in range(0,18,2)])
plt.ylabel('States')
plt.xlabel('Literacy Rate')
plt.title('Literacy Rate Increase in Urbans')
plt.show()
```

```python
plt.figure(figsize= (13,7))
sns.violinplot(ruralIncreases, color = 'red', palette = 'copper')
sns.violinplot(urbanIncreases, color = 'blue', palette = 'winter')
```

So there's a much larger variational increase in Literacy in the Rural Areas than in Urban Areas.
