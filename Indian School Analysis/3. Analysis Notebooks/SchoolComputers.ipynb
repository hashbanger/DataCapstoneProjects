```python
plt.figure(figsize= (20,20))
plt.subplot(5,2,1)
sns.barplot(df['State_UT'], df['Primary_Only'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,2)
sns.barplot(df['State_UT'], df['Primary_with_U_Primary'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,3)
sns.barplot(df['State_UT'], df['Primary_with_U_Primary_Sec'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,4)
sns.barplot(df['State_UT'], df['Primary_with_U_Primary_Sec_HrSec'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,5)
sns.barplot(df['State_UT'], df['U_Primary_Only'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,6)
sns.barplot(df['State_UT'], df['U_Primary_With_Sec'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,7)
sns.barplot(df['State_UT'], df['U_Primary_With_Sec_HrSec'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,8)
sns.barplot(df['State_UT'], df['Sec_Only'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,9)
sns.barplot(df['State_UT'], df['Sec_with_HrSec.'], palette = 'bone')
plt.xticks(rotation = 90)
plt.subplot(5,2,10)
sns.barplot(df['State_UT'], df['HrSec_Only'], palette = 'bone')
plt.xticks(rotation = 90)
plt.show()
```

#### The Average Percentage of Computers in combining all schools

```python
temp = pd.DataFrame(df.groupby(['State_UT']).mean().sort_values(by = 'All Schools')['All Schools']).reset_index()
temp = temp[temp['All Schools'] >= 50]
plt.figure(figsize= (15,15))
plt.subplot(2,1,1)
sns.barplot(y = temp['State_UT'], x = temp['All Schools'], palette = 'summer_r')
plt.xticks([50,60,70,80,90,100])
plt.title("Computers in above 50% schools", fontdict  ={'fontsize': 14})

temp2 = pd.DataFrame(df.groupby(['State_UT']).mean().sort_values(by = 'All Schools')['All Schools']).reset_index()
temp2 = temp2[temp2['All Schools'] <= 30]
plt.subplot(2,1,2)
sns.barplot(y = temp2['State_UT'], x = temp2['All Schools'], palette = 'autumn')
plt.xticks([10,20,30])
plt.title("Computers in below 30% schools", fontdict  ={'fontsize': 14})
plt.tight_layout()
plt.show()
```
