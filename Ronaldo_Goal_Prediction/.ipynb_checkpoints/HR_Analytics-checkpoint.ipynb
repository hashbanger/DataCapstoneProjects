## The Imports

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.offline as pyo
import plotly.graph_objs as go

import plotly.figure_factory as ff

from plotly import tools
py.offline.init_notebook_mode(connected= True)
from warnings import filterwarnings

filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

#### Checking for missing values.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), cbar = False)
plt.show()

# for plotting this with plotly
# data = [go.Heatmap(z = np.array(df.isna().replace([False, True], [0,1])))]
# py.offline.iplot(data)
```

We can observe that the missing values lie in the columns **education** and **previous_year_rating**.  
Let us check the sparsity of the columns to get the idea what to do with them.

```python
print('The sparsity of education colum is {}\nThe sparsity of previous_year_rating is {}'.format(df['education'].isna().sum()/len(df), df['previous_year_rating'].isna().sum()/len(df) ))
```

So we can see that the sparsity for both the columns is 5% and 7% so we can try imputing the values.

```python
list(df['education'].unique())
```

```python
df['education'].value_counts()
```

```python
df['education'] = df['education'].fillna(df['education'].value_counts().index[0])
df['education'].isna().sum()
```

```python
df['previous_year_rating'].unique()
```

Our second column **previous_year_rating** has float vlues but which are categorical and ordinal in nature.  
We can choose to fill it with most frequent value in the column or,  
we can impute it with the mean however it would make our column continuous in nature.  

```python
df['previous_year_rating'].value_counts()
```

```python
df['previous_year_rating'] = df['previous_year_rating'].fillna(df['previous_year_rating'].value_counts().index[0])
df['previous_year_rating'].isna().sum()
```

## EDA

Now we go through each column and explore it

### Department

```python
plt.figure(figsize=(14, 21))
plt.subplot(3,1,1)
sns.countplot(df['department'], order = df['department'].value_counts().index, palette= 'viridis')
plt.subplot(3,1,2)
sns.countplot(df['department'], order = df['department'].value_counts().index, hue= df['is_promoted'])
plt.subplot(3,1,3)
sns.countplot(df['department'], order = df['department'].value_counts().index, palette = 'Set1',hue= df['gender'])
plt.show()
```

**Observations**  
* We can observe that the most of the employees in the data are in the *Sales & Marketing* dept.  
* However the fraction of the employees getting promoted from the department is comparitively less to other departments.
* There are more fraction of males in each department but the ratio of females is very less in departments like Analytics, Legal and R&D.

### Education

```python
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
sns.countplot(y = df['education'], order = df['education'].value_counts().index, palette= 'Set2')
plt.title('Count of the Employees Education')
plt.subplot(1,2,2)
sns.countplot(y = df['education'], order = df['education'].value_counts().index, hue= df['is_promoted'],)
plt.yticks([])
plt.title('As per the promotion')
plt.ylabel('')
plt.show()
```

Again the number of bachelor employees in the department is more than twice the employees with masters and several times more than below secondary employees.

### Region

```python
df['region'].unique()
```

The **Region** contain region numbers along with 'region' as prefix. So we better slice the values to the numbers and convert it to integers.

```python
df['region'] = df['region'].apply(lambda x: int(x.split('_')[1]))
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(df['region'], palette = 'rainbow', order = np.sort(list(df['region'].unique())))
plt.show()
```

#### Observation  
There are comparatively more employees in the region or coming from the region 2, 7 and 22.

### Gender

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.countplot(df['gender'].replace(['m','f'], ['Male','Female']), palette= 'RdBu' )
plt.subplot(1,2,2)
plt.pie(df['gender'].value_counts(), labels = ['Male','Female'], autopct='%1.1f%%', colors= ['#4d94ff','#ff8080'], shadow= True)
plt.show()
```

As we have observed previously as well that the number of males in the employees is more than twice in the data.

### Recruitment Channel & Number of trainings

```python
plt.figure(figsize=(10, 12))
plt.subplot(2,1,1)
sns.countplot(df['recruitment_channel'], palette= 'Wistia', order = df['recruitment_channel'].value_counts().index)
plt.subplot(2,1,2)
sns.countplot(df['no_of_trainings'], palette= 'cool', order = df['no_of_trainings'].value_counts().index)
plt.show()
```

**Observations**    
* We can observe that most of the recruitments are from other sources and a good amount from sourcing.   
* Most of the employees have underwent only 1 training

### Age

```python
plt.figure(figsize=(14,12))

plt.subplot(2,1,1)
plt.hist(df['age'])
plt.xticks(list(range(0,60,5)))
plt.subplot(2,1,2)
sns.distplot(df[df['gender'] == 'm']['age'], label= 'Male')
sns.distplot(df[df['gender'] == 'f']['age'],label= 'Female')
plt.legend()
plt.xticks(list(range(0,60,5)))
plt.show()
```

**Observations**  
* It can be observed that most of the HRs are in the age group 25-40.  
* It is evident that there is no difference between the age groups of the employees on basis of the gender.

```python
plt.figure(figsize=(21,6))
plt.subplot(1,3,1)
sns.regplot(df['age'], df['no_of_trainings'], color = 'green')
plt.subplot(1,3,2)
sns.regplot(df['age'], df['previous_year_rating'], color = 'blue')
plt.subplot(1,3,3)
sns.regplot(df['age'], df['length_of_service'], color = 'orange')
plt.show()
```

**Observation**  
* The number of trainings are not varying as per the age.  
* The previous year rating is also not a function of the age.
* This is a significant relationship that the length of service has a linear relation with age which makes sense as well.

### Previous Year Rating

```python
plt.figure(figsize=(12, 6))
sns.countplot(df['previous_year_rating'], palette = 'pastel')
plt.title('Most Frequent Ratings')
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.countplot(df['previous_year_rating'], palette = 'cool', hue= df['gender'])
plt.title('Rating as per gender')
plt.subplot(1,2,2)
sns.barplot(x = [',female','male'], y = [df[df['gender'] == 'f']['previous_year_rating'].mean(),df[df['gender'] == 'm']['previous_year_rating'].mean()], palette= 'cool')
plt.title('Avg. Rating as per gender')
plt.show()
```

We can see that the average ratings among the males and females is almost equal.

### KPIs_met>80% and Awards won
